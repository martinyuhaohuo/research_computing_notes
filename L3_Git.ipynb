## Lecture 3: Git

Git is a distributed version control system that tracks versions of files. It is used to control source code by programmers who are developing software collaboratively.

Git repositories can be hosted both locally and remotelly; there main platform for hosting remote repositories are:
1. Github
2. Gitlab
3. BitBucket

We can connect local repo to remote repo, enabling remote backup and collobration.

### Collobration Workflow using Github and Git

#### Fork
If we are working on a project already hosted on github we should fork it to our account. 
This will allow us to modify it, keep track of what we are doing, and contribute to it.

#### Clone
- We can clone the forked github repository to our account with the git clone command.
- Or if we start the project, we can create the github repo at first and connect it to the local repo by cloning it.
- It is achived through `git clone https://...`

#### Pull
- When multiple people are collaborating on a same repo, it is possible that the local repo is out of date because someone else has made an update.
- We can first run `git fetch` and then use `git status` to check the status of our local repo.
- If the above command show that the local repo is behind the remote repo, we can update it with `git pull`.
- Otherwise, we can run git pull every time before we are going to work on the code.

#### Commit
- `git status` check uncommitted changes and untracked files
- `git log` check commit history
- `git add .` add all files to stage before commit
- `git commit -m "message"` commit the changes, track of what we have done
- `git push` push the change to the remote repo

#### Branches
- A git repository can have many branches which intend to develop different features of the project. 
- At some point, changes from different branches can be merged back to the main (complete the development of new feature).
- first we create a new branch `git switch -c <new_branch>`
- modifying code on it
- save changes, add changes, and commit changes
- after review, we may decide to merge the branch back to main by `git switch main` then `git merge <new_branch>`

#### Pull Request
- Pull requests are used to merge changes from our newly created branch into the main branch of the remote repository (if we are coauthors).
- Or when we want to merge a forked branch to the original main (if we are not coauthors).

#### Overall Collobration Workflow
- You are a collaborator
1. git clone the repo `git clone <repo-url>`
2. create a new local branch `git checkout -b feature-xyz`
3. edit the code of on the new branch to add feature
4. git add . and git commit changes
5. git push the new feature branch to github remote repo `git push -u origin feature-xyz`
6. open a pull request for merging the new branch to the origin/main
7. peer review and approve
- Not a collaborator
1. fork the origin repo
2. git clone
3. make changes on local repo
4. git add., git commit, and git push
5. submit a pull request
